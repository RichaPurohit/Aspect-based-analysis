````python
import pandas as pd
import nltk
import string
import matplotlib.pyplot as plt
import seaborn as sns
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
````

````python
# Setup Chrome options
options = Options()
options.add_argument("--headless")  # Run headless browser for silent scraping
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Initialize the WebDriver using ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
````

````python
def search_amazon_brand(brand_name):
    base_url = 'https://www.amazon.in'
    url = f'{base_url}/s?k={brand_name}'
    driver.get(url)
    
    products = []
    
    while True:
        # Wait for the page to load and products to be present
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//div[@data-component-type="s-search-result"]'))
        )
        
        # Extract product details
        product_elements = driver.find_elements(By.XPATH, '//div[@data-component-type="s-search-result"]')
        for product_element in product_elements:
            try:
                # Extracting title
                title_element = product_element.find_element(By.XPATH, './/span[contains(@class, "a-size-medium") or contains(@class, "a-size-base-plus")]')
                title = title_element.text
                
                # Extracting link
                link_element = product_element.find_element(By.XPATH, './/a[@class="a-link-normal s-no-outline"]')
                link = link_element.get_attribute('href')
                
                # Extracting price
                price = "N/A"
                try:
                    price_whole = product_element.find_element(By.XPATH, './/span[@class="a-price-whole"]').text
                    price_fraction = product_element.find_element(By.XPATH, './/span[@class="a-price-fraction"]').text if len(product_element.find_elements(By.XPATH, './/span[@class="a-price-fraction"]')) > 0 else '00'
                    price = f"₹{price_whole}.{price_fraction}"
                except:
                    try:
                        price = product_element.find_element(By.XPATH, './/span[@class="a-offscreen"]').text
                    except:
                        pass
                
                # Extracting rating
                rating = "N/A"
                try:
                    rating_element = product_element.find_element(By.XPATH, './/span[@class="a-icon-alt"]')
                    rating = rating_element.get_attribute('textContent')
                except:
                    pass
                
                # Extracting review count
                review_count = "N/A"
                try:
                    review_count_element = product_element.find_element(By.XPATH, './/span[@class="a-size-base"]')
                    review_count = review_count_element.text
                except:
                    pass
                
                
                # Append product details to the list
                products.append({'title': title, 'link': link, 'price': price, 'rating': rating, 'review_count': review_count, })
            except Exception as e:
                print(f"Error extracting product: {e}")
                continue
        
        # Go to the next page if available
        try:
            next_page = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//li[@class="a-last"]/a'))
            )
            next_page.click()
        except Exception as e:
            print(f"Error navigating to next page: {e}")
            break
    
    return pd.DataFrame(products)

# List of brand names to extract products for
brand_names = ['Janasya', 'Seryeon', 'Azira','Lymio','AASK','SIRIL','POSHAX','Sugathari','Blushia','Styli','Preneum','Miss Chase']
all_products = []

for brand_name in brand_names:
    print(f"Extracting products for brand: {brand_name}")
    brand_products_df = search_amazon_brand(brand_name)
    all_products.append(brand_products_df)

# Combine all brand products into a single DataFrame
combined_products_df = pd.concat(all_products, ignore_index=True)
````

````python
def search_amazon_reviews(product_url, product_identifier):
    driver.get(product_url)
    
    reviews = []
    
    # Click "See more reviews" if available
    try:
        see_more_reviews = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//a[@data-hook="see-all-reviews-link-foot"]'))
        )
        see_more_reviews.click()
    except Exception as e:
        print(f"Error clicking 'See more reviews': {e}")

    while True:
        # Wait for the page to load and reviews to be present
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//div[@data-hook="review"]'))
            )
        except Exception as e:
            print(f"Error waiting for reviews to load: {e}")
            break
        
        # Extract review details
        review_elements = driver.find_elements(By.XPATH, '//div[@data-hook="review"]')
        for review_element in review_elements:
            try:
                review_title = review_element.find_element(By.XPATH, './/a[@data-hook="review-title"]').text
                review_date = review_element.find_element(By.XPATH, './/span[@data-hook="review-date"]').text
                review_body = review_element.find_element(By.XPATH, './/span[@data-hook="review-body"]').text
                
                reviews.append({
                    'product_identifier': product_identifier,
                    'review_title': review_title,
                    'review_date': review_date,
                    'review_body': review_body
                })
            except Exception as e:
                print(f"Error extracting review: {e}")
                continue
        
        # Go to the next page if available
        try:
            next_page_button = driver.find_element(By.XPATH, '//li[@class="a-last"]/a')
            if 'a-disabled' in next_page_button.get_attribute('class'):
                print("No more pages left")
                break
            next_page_button.click()
            
            # Wait for the next page to load
            WebDriverWait(driver, 10).until(
                EC.staleness_of(review_elements[0])
            )
        except Exception as e:
            print(f"Error navigating to next page: {e}")
            break
    
    return reviews
````

````python
products_df = pd.read_csv('all_brands_products.csv')  #CSV with product details
all_reviews = []

for index, row in products_df.iterrows():
    product_identifier = row['link']  # Use 'link' as the unique identifier
    product_url = row['link']
    print(f"Extracting reviews for {product_identifier}")
    product_reviews = search_amazon_reviews(product_url, product_identifier)
    all_reviews.extend(product_reviews)
````

## Text Preprocessing and Cleaning

````python
# Check for missing values
print(merged_df.isnull().sum())
````

````python
# Handle missing values
merged_df['review_title'].fillna('No Title', inplace=True)
merged_df['review_body'].fillna('No Review', inplace=True)
merged_df['price'].fillna('No Price', inplace=True)

# Verify that there are no missing values left
print(merged_df.isnull().sum())
````

````python
# Extract the date from the 'review_date' column
merged_df['review_date'] = merged_df['review_date'].str.extract(r'on (\d+ \w+ \d{4})')[0]

# Convert the extracted date to datetime format
merged_df['review_date'] = pd.to_datetime(merged_df['review_date'], format='%d %B %Y')

# Extract the numeric part of the rating
merged_df['rating'] = merged_df['rating'].str.extract(r'(\d\.\d)')[0]

# Convert the extracted string to a float
merged_df['rating'] = merged_df['rating'].astype(float)

# Clean the price column by removing the currency symbol and converting to float
merged_df['price'] = merged_df['price'].str.replace('₹', '').str.replace(',', '')
merged_df['price'] = pd.to_numeric(merged_df['price'], errors='coerce')
````

````python
# Text cleaning with negation handling
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
negation_words = set(["not", "no", "never", "none", "nobody", "nothing", "neither", "nor", "nowhere", "hardly", "scarcely", "barely", "don't", "isn't", "wasn't", "shouldn't", "wouldn't", "couldn't", "won't", "can't", "doesn't"])

def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    clean_words = []
    negation = False
    for word in words:
        if word in negation_words:
            negation = True
            clean_words.append(word)
        elif negation:
            clean_words.append('NOT_' + word)
            negation = False
        elif word not in stop_words:
            clean_words.append(word)
    return TreebankWordDetokenizer().detokenize(clean_words)

# Clean the review_body column
merged_df['cleaned_review_body'] = merged_df['review_body'].apply(clean_text)
merged_df['cleaned_review_title'] = merged_df['review_title'].apply(clean_text)
````

````python
# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to calculate sentiment scores
def get_sentiment_score(review):
    sentiment = analyzer.polarity_scores(review)
    return sentiment['compound']

# Apply sentiment analysis to the cleaned review body
merged_df['sentiment_score'] = merged_df['cleaned_review_body'].apply(get_sentiment_score)

# Group by product_identifier and calculate aggregate metrics
grouped_df = merged_df.groupby('product_identifier').agg({
    'review_title': 'count',
    'review_body': lambda x: ' '.join(x),
    'title': 'first',
    'price': 'first',
    'rating': 'mean',
    'sentiment_score': 'mean'
}).reset_index()

# Rename the columns for clarity
grouped_df.rename(columns={
    'review_title': 'review_count',
    'review_body': 'all_reviews',
    'rating': 'average_rating',
    'sentiment_score': 'average_sentiment_score'
}, inplace=True)
````

````python
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part, FinishReason
import vertexai.preview.generative_models as generative_models
from google.cloud import aiplatform
import json
import time
import pandas as pd
````

````python
PROJECT_ID = "vertexaiproject-432115"  # @param {type:"string"}
REGION = "us-central1"  # @param {type: "string"}
generation_config = {
    "max_output_tokens": 2048,
    "temperature": 1,
    "top_p": 0.4,
    "top_k": 32,
}
api_endpoint = 'us-central1-aiplatform.googleapis.com'
client_options = {"api_endpoint": api_endpoint}

safety_settings = {
    generative_models.HarmCategory.HARM_CATEGORY_HATE_SPEECH: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_HARASSMENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
}
````

````python
# Add new columns for the results
merged_df['polarity'] = ""
merged_df['aspect_based_opinion'] = ""
merged_df['things_to_improve'] = ""

# File path to save progress
output_file_path = 'merged_df_sample_progress.csv'

# Function to generate responses from the model
def generate(text1):
    vertexai.init(project=PROJECT_ID, location=REGION)
    model = GenerativeModel("gemini-1.5-pro-preview-0409")
    responses = model.generate_content(
        text1,
        generation_config=generation_config,
        safety_settings=safety_settings,
        stream=True,
    )
    text_output = ""
    for response in responses:
        text_output += response.text
    return text_output

# Function to process each review and extract the necessary details
def code_score(review):
    query = f"""
    Help me know the polarity/sentiment and aspect based opinion(in one line summary), things to improve(keywords) of following review:

    {review}
    
    Additionally, ensure that the output follows this JSON format:
      {{
          "Polarity": ,
          "Aspect based opinion": ,
          "Things to improve":
      }}
    """
    query_output = generate(query)
    query_output_f1 = query_output.replace('```', "").replace("json", "").replace("\n", "")
    score_data = json.loads(query_output_f1)
    return score_data

# Loop through all reviews and apply the model one by one with error handling
def process_reviews_one_by_one(df, delay=10, save_interval=1):
    results = []  # To store the results of each review
    
    for idx, review in enumerate(df['review_title']+' '+df['review_body']):
        try:
            print(f"Processing review {idx + 1}/{len(df)}")
            result = code_score(review)
            results.append(result)

            # Store the results in the dataframe
            df.at[idx, 'polarity'] = result.get('Polarity', "")
            df.at[idx, 'aspect_based_opinion'] = result.get('Aspect based opinion', "")
            df.at[idx, 'things_to_improve'] = result.get('Things to improve', "")

            print(f"Result for review {idx + 1}: {result}")
            
        except Exception as e:
            print(f"Error processing review {idx + 1}: {e}")
            results.append(f"Error processing review {idx + 1}")
        
        # Save progress after every 'save_interval' reviews
        if (idx + 1) % save_interval == 0:
            df.to_csv(output_file_path, index=False)  # Save the DataFrame to a CSV file
            print(f"Progress saved to {output_file_path} after processing {idx + 1} reviews.")
        
        # Add a delay between requests to avoid quota issues
        if idx < len(df) - 1:  # No need to delay after the last review
            print(f"Waiting for {delay} seconds before processing the next review...")
            time.sleep(delay)  # Add delay between each API call
    
    # Final save after all reviews are processed
    df.to_csv(output_file_path, index=False)
    print(f"Final progress saved to {output_file_path}.")
    
    return df

# Run the process on all reviews and update the DataFrame
merged_df = process_reviews_one_by_one(merged_df, save_interval=100)  # Adjust delay and save_interval as needed

# Now merged_df_sample contains the polarity, aspect_based_opinion, and things_to_improve columns
print(merged_df.head())  # Display the first few rows of the updated dataframe
````

````python
# Function to clean and process 'things_to_improve' entries
def clean_things_to_improve(entry):
    if isinstance(entry, str):
        # Convert string representations of lists into actual lists
        entry = entry.replace('[', '').replace(']', '').replace('\'', '').split(',')
        # Strip whitespace and remove empty strings
        entry = [item.strip() for item in entry if item.strip()]
    elif isinstance(entry, list):
        # If already a list, clean the items inside the list
        entry = [item.strip() for item in entry if item.strip()]
    else:
        entry = []  # Handle any other case as an empty list

    return entry

# Apply the cleaning function to the 'things_to_improve' column
output_sample['things_to_improve'] = output_sample['things_to_improve'].apply(clean_things_to_improve)

# Group the cleaned data
grouped_data = output_sample.groupby('product_identifier').agg({
    'polarity': lambda x: x.value_counts().index[0] if not x.value_counts().empty else None,  # Most common polarity
    'things_to_improve': lambda x: [item for sublist in x for item in sublist if item]  # Flatten and remove empty entries
}).reset_index()
````
